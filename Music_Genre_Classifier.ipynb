```python
#improting the tools
import librosa
import numpy as np
from datasets import load_dataset
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
# Download the Music Data
print("Downloading music...")
gtzan = load_dataset("mteb/gtzan-genre", split="train")
print(f"Successfully downloaded {len(gtzan)} songs.")
```

```python
# Extract Features from the Audio
features = []
labels = []

# The dataset uses 'label' instead of 'genre'
print("Extracting audio features. Please wait...")
num_samples = min(100, len(gtzan))

for i in range(num_samples):
    song = gtzan[i]
    audio_array = song["audio"]["array"]
    sample_rate = song["audio"]["sampling_rate"]
    label = song["label"]

    # Extract MFCCs
    mfccs = librosa.feature.mfcc(y=audio_array, sr=sample_rate, n_mfcc=13)

    # Calculate the average of the MFCCs
    mfccs_average = np.mean(mfccs, axis=1)

    features.append(mfccs_average)
    labels.append(label)

print(f"Extraction complete for {num_samples} samples.")
```

```python
# Train the Machine Learning Model
```

```python
# Split data: 80% for training, 20% for testing
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)

# Create the AI model
ai_model = KNeighborsClassifier(n_neighbors=3)

# Train the model
ai_model.fit(X_train, y_train)
print("Model training complete.")
```

```python
# Test the Model
```

```python
predictions = ai_model.predict(X_test)

# Calculate the score
score = accuracy_score(y_test, predictions)
print(f"The AI correctly guessed: {score * 100}% of the songs.")
```
